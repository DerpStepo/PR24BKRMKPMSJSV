# Kakovost življenja v EU
## Projektna naloga - podatkovno rudarjenje 

# UVOD 
Pri projektu smo analizirali oceno kakovosti življenja prebivalcev različnih držav Evropske unije. Zanimalo nas kateri faktorji (npr. BDP na prebivalca države, kvaliteta šolskega sistema, mere neenakosti, občutek varnosti prebivalca, ure dela tedensko oz. work-life balance itd.) najmočneje vplivajo na kakovost življenja posameznika.

# PODATKI

Za pridobivanje podatkov smo uporabili spletne strani, ki so namenjene za hranjenje statistik in informacij. Glavna stran, ki smo jo uporabili je bila Eurostat, s kjer smo pridobili 9 CVS datotek. Uporabili pa smo tudi Worldeconomics kjer smo pridobili eno CSV datoteko (EqualityIndex) ter Numbeo, kjer smo ravno tako pridobili eno datoteko (QOL_Index).

Pridobljeni podatki so:
1. Indesk enakovrednosti med spoloma,
2. Tedenske delovne ure,
3. Infacija,
4. Brezposelnost (na 1000 prebivalcev),
5. Migracije, 
6. Emigracije, 
7. Minimalna plača, 
8. Delovne ure (na teden),
9. Bruto domači proizvod (BDP) 
10. Kvaliteta življenja (QOL) in 
11. Procent oseb ki obupajo nad šolanjem.

Podatki so večinoma ločeni na države in leta, ko so bile mere opravljene. Mere se delijo na povprečje, procentualno na populacijo in samo prešteti elementi (npr. število oseb).

# VPRAŠANJA / CILJI
Naš cilj pri obdelavi podatkov in izdelavi grafov je bil odgovoriti na vprašanja:

1. Korelacija med inflacijo in GDP/Minimalno plačo​
2. Ali odstotek nezaposlene populacije vpliva na GDP​
3. Kakšne bodo potencialne inflacije v naprej, glede na dobljene podatke (višje/nižje)​
